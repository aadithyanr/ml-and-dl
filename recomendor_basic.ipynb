```python
import pandas as pd
```

```python
r_cols = ['user_id', 'movie_id', 'rating'] 
ratings = pd.read_csv(r'C:\Users\aadithyan.NAIR\Desktop\Machine Learning\ml-100k\u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv(r'C:\Users\aadithyan.NAIR\Desktop\Machine Learning\ml-100k\u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")
```

```python
movies.head()
```

```python
ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python
tilltherewas = movieRatings['12 Angry Men (1957)']
tilltherewas.head()
```

```python
similarMovies = movieRatings.corrwith(tilltherewas)
similarMovies = similarMovies.dropna()
similarMovies.head()
```

```python
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

## pretty basic example of a reccomendation per movie

now, let's scale it on a general level. let's say we want to recommend movies to a user based on their movie ratings. we can use the same approach as before, but now we will use the ratings of the movies as the features for the recommendation.

```python
import numpy as np
movieStats = ratings.groupby(['title']).agg({'rating': [np.size, np.mean]})
movieStats.head(5)
```

SMASSSHHHHHH. NOW LETS REMOVE THE LOW RATED MOVIES AND ONLY KEEP THE HIGH RATED MOVIES

```python
popularStats = movieStats['rating']['size'] >= 100
popularStats.head()
```

```python
movieStats[popularStats].sort_values([('rating', 'mean')], ascending=False) [:15]
```

```python
movieStats.columns = movieStats.columns.get_level_values(0)  # flattening the multi-index columns
df = movieStats[popularStats].join(pd.DataFrame(similarMovies, columns=['similarity']), how='inner')
```
